```python
import gmsh
import os
import numpy as np
import sys

lc = 2.5e-3; lc_set = 5e-3; z_coor = 0; scale = 1000;

def game_on_bell_case():
    
    gmsh.initialize()
    gmsh.model.add("bell_case")

    gmsh.model.geo.addPoint(30/scale, 150/scale, z_coor, lc, 1)
    gmsh.model.geo.addPoint(30/scale, 190.4/scale, z_coor, lc, 2)
    gmsh.model.geo.addPoint(130/scale, 150/scale, z_coor, lc, 3)
    gmsh.model.geo.addPoint(30/scale, 185.2/scale, z_coor, lc, 4)
    gmsh.model.geo.addPoint(19.9/scale, 200.9/scale, z_coor, lc, 5)
    gmsh.model.geo.addPoint(71/scale, 235.4/scale, z_coor, lc, 6)
    gmsh.model.geo.addPoint(77/scale, 235.4/scale, z_coor, lc, 7)
    gmsh.model.geo.addPoint(-200/scale, 200.9/scale, z_coor, lc, 8)
    gmsh.model.geo.addPoint(130/scale, 450/scale, z_coor, lc, 9)
    gmsh.model.geo.addPoint(-200/scale, 205.9/scale, z_coor, lc, 10)
    gmsh.model.geo.addPoint(19.9/scale, 205.9/scale, z_coor, lc, 11)
    gmsh.model.geo.addPoint(66/scale, 235.4/scale, z_coor, lc, 12)
    gmsh.model.geo.addPoint(82/scale, 235.4/scale, z_coor, lc, 13)
    gmsh.model.geo.addPoint(-200/scale, 450/scale, z_coor, lc, 14)
    gmsh.model.geo.addPoint(42.3/scale, 191.6/scale, z_coor, lc, 15)
    gmsh.model.geo.addPoint(48.7/scale, 193.6/scale, z_coor, lc, 16)
    gmsh.model.geo.addPoint(55.2/scale, 196.8/scale, z_coor, lc, 17)
    gmsh.model.geo.addPoint(62.3/scale, 202.1/scale, z_coor, lc, 18)
    gmsh.model.geo.addPoint(67.3/scale, 207.4/scale, z_coor, lc, 19)
    gmsh.model.geo.addPoint(71/scale, 212.9/scale, z_coor, lc, 20)
    gmsh.model.geo.addPoint(73.7/scale, 218.4/scale, z_coor, lc, 21)
    gmsh.model.geo.addPoint(75.7/scale, 224.8/scale, z_coor, lc, 22)
    gmsh.model.geo.addPoint(76.8/scale, 231.4/scale, z_coor, lc, 23)
    gmsh.model.geo.addPoint(76.9/scale, 236.1/scale, z_coor, lc, 24)
    gmsh.model.geo.addPoint(76.4/scale, 237.2/scale, z_coor, lc, 25)
    gmsh.model.geo.addPoint(75.8/scale, 237.8/scale, z_coor, lc, 26)
    gmsh.model.geo.addPoint(74.4/scale, 238.4/scale, z_coor, lc, 27)
    gmsh.model.geo.addPoint(73.2/scale, 238.3/scale, z_coor, lc, 28)
    gmsh.model.geo.addPoint(72.2/scale, 237.8/scale, z_coor, lc, 29)
    gmsh.model.geo.addPoint(71.5/scale, 237.1/scale, z_coor, lc, 30)
    gmsh.model.geo.addPoint(71.1/scale, 236.2/scale, z_coor, lc, 31)
    gmsh.model.geo.addPoint(31/scale, 202.4/scale, z_coor, lc, 32)
    gmsh.model.geo.addPoint(43/scale, 206.1/scale, z_coor, lc, 33)
    gmsh.model.geo.addPoint(53.4/scale, 211.9/scale, z_coor, lc, 34)
    gmsh.model.geo.addPoint(62.8/scale, 220.3/scale, z_coor, lc, 35)
    gmsh.model.geo.addPoint(67.3/scale, 226.2/scale, z_coor, lc, 36)
    gmsh.model.geo.addPoint(69.9/scale, 231.3/scale, z_coor, lc, 37)

    gmsh.model.geo.addLine(1, 3, 38)
    gmsh.model.geo.addLine(9, 14, 39)
    gmsh.model.geo.addLine(8, 5, 40)
    gmsh.model.geo.addLine(10, 11, 41)
    gmsh.model.geo.addLine(8, 10, 42)
    gmsh.model.geo.addLine(5, 11, 43)
    gmsh.model.geo.addLine(10, 14, 44)
    gmsh.model.geo.addLine(1, 4, 45)
    gmsh.model.geo.addLine(4, 2, 46)
    gmsh.model.geo.addLine(6, 12, 47)
    gmsh.model.geo.addLine(7, 13, 48)
    gmsh.model.geo.addLine(3, 9, 49)

    gmsh.model.geo.addSpline([2, 15, 16, 17, 18, 19, 20, 21, 22, 23, 7], 50)
    gmsh.model.geo.addSpline([7, 24, 25, 26, 27, 28, 29, 30, 31, 6], 51)
    gmsh.model.geo.addSpline([6, 37, 36, 35, 34, 33, 32, 5], 52)

    gmsh.model.geo.addPoint(74/scale, 234.4/scale, z_coor, lc, 53)
    gmsh.model.geo.addPoint(32/1000, 235.4/1000, z_coor, lc, 54)
    gmsh.model.geo.addPoint(9.2/1000, 273.4/1000, z_coor, lc, 55)

    gmsh.model.geo.addCircleArc(11, 55, 12, 56)
    gmsh.model.geo.addCircleArc(4, 54, 13, 57)
    gmsh.model.geo.addCircleArc(12, 53, 13, 58)

    gmsh.model.geo.synchronize()

    gmsh.model.geo.addCurveLoop([42, 41, -43, -40], 59)
    gmsh.model.geo.addCurveLoop([43, 56, -47, 52], 60)
    gmsh.model.geo.addCurveLoop([47, 58, -48, 51], 61)
    gmsh.model.geo.addCurveLoop([48, -57, 46, 50], 62)
    gmsh.model.geo.addCurveLoop([44, -39, -49, -38, 45, 57, -58, -56, -41], 63)

    gmsh.model.geo.addPlaneSurface([59], 64)
    gmsh.model.geo.addPlaneSurface([60], 65)
    gmsh.model.geo.addPlaneSurface([61], 66)
    gmsh.model.geo.addPlaneSurface([62], 67)
    gmsh.model.geo.addPlaneSurface([63], 68)

    x = 1.02

    for l in [51, 58]:
        gmsh.model.geo.mesh.setTransfiniteCurve(l, 30, "Progression", 1.0)
    for l in [57, 50]:
        gmsh.model.geo.mesh.setTransfiniteCurve(l, 75, "Progression", 0.99)
    for l in [52]:
        gmsh.model.geo.mesh.setTransfiniteCurve(l, 50, "Progression", x)
    for l in [56]:
        gmsh.model.geo.mesh.setTransfiniteCurve(l, 50, "Progression", 1/x)
    for l in [47, 43, 42, 48]:
        gmsh.model.geo.mesh.setTransfiniteCurve(l, 30, "Progression", 1.1)
    for l in [40, 41]:
        gmsh.model.geo.mesh.setTransfiniteCurve(l, 125, "Progression", 1)
    for l in [46]:
        gmsh.model.geo.mesh.setTransfiniteCurve(l, 30, "Progression", 0.9)


    gmsh.model.geo.mesh.setTransfiniteSurface(67)
    gmsh.model.geo.mesh.setTransfiniteSurface(66)
    gmsh.model.geo.mesh.setTransfiniteSurface(65)
    gmsh.model.geo.mesh.setTransfiniteSurface(64)

    gmsh.model.geo.mesh.setRecombine(2, 67)
    gmsh.model.geo.mesh.setRecombine(2, 66)
    gmsh.model.geo.mesh.setRecombine(2, 65)
    gmsh.model.geo.mesh.setRecombine(2, 64)

    gmsh.model.mesh.field.add("Box", 1)
    gmsh.model.mesh.field.setNumber(1, "VIn", lc/3 )
    gmsh.model.mesh.field.setNumber(1, "VOut", lc)
    gmsh.model.mesh.field.setNumber(1, "XMin", 0.03) 
    gmsh.model.mesh.field.setNumber(1, "XMax", 0.04)
    gmsh.model.mesh.field.setNumber(1, "YMin", 0.15)
    gmsh.model.mesh.field.setNumber(1, "YMax", 0.185)

    gmsh.model.mesh.field.add("Distance", 2)
    gmsh.model.mesh.field.setNumbers(2, "CurvesList", [45])
    gmsh.model.mesh.field.setNumber(2, "Sampling", 25)

    gmsh.model.mesh.field.add("Threshold", 3)
    gmsh.model.mesh.field.setNumber(3, "InField", 2)
    gmsh.model.mesh.field.setNumber(3, "SizeMin", lc / 4)
    gmsh.model.mesh.field.setNumber(3, "SizeMax", lc)
    gmsh.model.mesh.field.setNumber(3, "DistMin", 0.01)
    gmsh.model.mesh.field.setNumber(3, "DistMax", 0.22/2)

    gmsh.model.mesh.field.add("Min", 4)
    gmsh.model.mesh.field.setNumbers(4, "FieldsList", [1, 3])
    gmsh.model.mesh.field.setAsBackgroundMesh(4)


    gmsh.option.setNumber("Mesh.MeshSizeExtendFromBoundary", 0)
    gmsh.option.setNumber("Mesh.MeshSizeFromPoints", 0)
    gmsh.option.setNumber("Mesh.MeshSizeFromCurvature", 0)

    gmsh.model.geo.synchronize()

    # Extrusion
    layers = 1; extrude_height = 0.02; extrude_vector = [0, 0, extrude_height];
    ex = gmsh.model.geo.extrude([(2, 64), (2, 65), (2, 66), (2, 67), (2, 68)], extrude_vector[0], extrude_vector[1], extrude_vector[2], numElements = [layers], recombine = True)

    gmsh.model.geo.synchronize()

    # Physical Surfaces
    surface_ids = sorted([e[1] for e in ex if e[0] == 2])
    volume_ids = [e[1] for e in ex if e[0] == 3]

    # Print surface and volume IDs for verification
    # print("Extruded surface IDs: ", surface_ids)
    # print("Volume IDs: ", volume_ids)

    gmsh.model.addPhysicalGroup(3, volume_ids, 1)
    gmsh.model.setPhysicalName(3, 1, "Fluid")

    gmsh.model.addPhysicalGroup(2, [174, 178], 2)
    gmsh.model.setPhysicalName(2, 2, "Wall")
    gmsh.model.addPhysicalGroup(2, [89, 111, 133, 155], 3)
    gmsh.model.setPhysicalName(2, 3, "Bell")
    gmsh.model.addPhysicalGroup(2, [64, 65, 66, 67, 68, 90, 112, 134, 156, 203], 4)
    gmsh.model.setPhysicalName(2, 4, "Sides")
    gmsh.model.addPhysicalGroup(2, [170, 77], 5)
    gmsh.model.setPhysicalName(2, 5, "Inlet")
    gmsh.model.addPhysicalGroup(2, [186, 151], 6)
    gmsh.model.setPhysicalName(2, 6, "Outlet")
    gmsh.model.addPhysicalGroup(2, [182], 7)
    gmsh.model.setPhysicalName(2, 7, "Symmetry")




    # Mesh Generation
    gmsh.model.mesh.generate(3)

    mesh_file = os.path.abspath("game_on_bell_case_extra_fine.msh2")
    gmsh.write(mesh_file)

    # Launch the GUI to see the results:
    if '-nopopup' not in sys.argv:
        gmsh.fltk.run()

    gmsh.finalize()



if __name__ == "__main__":
    game_on_bell_case()
```
